```python
total_orders = pd.read_sql("""
    SELECT COUNT(DISTINCT [Order ID]) as Total_Orders,
    SUM(Quantity) as Total_Quantity
    FROM superstore
""", conn)
total_orders.to_csv('total_orders.csv', index=False)
```

```python
# Add this to your Colab notebook
sales_by_cat = pd.read_sql("SELECT Category, SUM(Sales) as Sales, SUM(Profit) as Profit FROM superstore GROUP BY Category", conn)
sales_by_market = pd.read_sql("SELECT Market, SUM(Sales) as Sales FROM superstore GROUP BY Market", conn)
monthly = pd.read_sql("SELECT strftime('%Y-%m', [Order Date]) as Month, SUM(Sales) as Sales FROM superstore GROUP BY Month", conn)

# Export
sales_by_cat.to_csv('sales_by_category.csv', index=False)
sales_by_market.to_csv('sales_by_market.csv', index=False)
monthly.to_csv('monthly_sales.csv', index=False)
```
